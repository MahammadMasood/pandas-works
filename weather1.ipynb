```python
import numpy  as numpyInstace
import pandas  as pandasInstance
import matplotlib.pyplot  as matplotlibInstace
import seaborn  as seabornInstance
```

```python
weather = pandasInstance.read_csv('weather.csv')
weather.info()
```

```python
weather.describe()
```

```python
seabornInstance.scatterplot(x='MinTemp', y='Humidity3pm', data=weather)
```

```python
seabornInstance.jointplot(x='MinTemp', y='Humidity3pm', data=weather,kind='hex')
```

```python
seabornInstance.lmplot(x='MinTemp', y='Humidity3pm', data=weather)
```

```python
X_features=weather[['Humidity3pm']]
X_features
```

```python
Y_Prediction=weather[['MinTemp']]
Y_Prediction
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_features, Y_Prediction, test_size=0.2, random_state=0)

# Create an instance of the LinearRegression model
linearRegression = LinearRegression()

# Fit the model to the training data
linearRegression.fit(X_train, y_train)

# Make predictions on the test data
temperaturePredictions = linearRegression.predict(X_test)
```

```python
temperaturePredictions
```

```python
matplotlibInstace.scatter(y_test, temperaturePredictions)
```

```python
seabornInstance.distplot(y_test-temperaturePredictions)
```

```python
from sklearn  import metrics
```

```python
metrics.mean_squared_error(y_test, temperaturePredictions)
```

```python
numpyInstace.sqrt(metrics.mean_squared_error(y_test, temperaturePredictions))
```
